```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('bechdelExpanded.csv')
print(df.info())

df['total_score'] = df['bechdel'] + df['ko'] + df['waithe']

df_sorted = df.sort_values("total_score").reset_index(drop = True)

df_sorted_final = df_sorted[['movie','bechdel','waithe','ko','total_score']]
print(df_sorted_final)

movie = df_sorted_final.movie
totalScore = df_sorted_final.total_score

x_ticks = list(range(0,50))
y_ticks = range(0,4)

plt.figure(figsize = (15,17))
az = plt.subplot(1,1,1)
plt.barh(x_ticks,totalScore)
plt.xlabel('Total Score',fontsize=13)
plt.ylabel('Movies',fontsize=13)
plt.title('Representation in Movies', fontsize=13)
az.set_yticks(x_ticks)
az.set_yticklabels(movie,fontsize=13,rotation = 0)
az.set_xticks(y_ticks)
plt.show()
```
